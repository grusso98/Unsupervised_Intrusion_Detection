## CPU metrics

```python
def get_cpu_metrics_name(mode):
  cpu_node3 = cpu.loc[:,cpu.columns[pd.Series(cpu.columns.values).apply(lambda x: 'instance="10.99.163.197:9100",job="node-exporter",mode="'+str(mode) + '"' in x)]]
  cpu_node2 = cpu.loc[:,cpu.columns[pd.Series(cpu.columns.values).apply(lambda x: 'instance="10.99.163.196:9100",job="node-exporter",mode="'+str(mode) + '"' in x)]]
  cpu_node1 = cpu.loc[:,cpu.columns[pd.Series(cpu.columns.values).apply(lambda x: 'instance="10.99.163.195:9100",job="node-exporter",mode="'+str(mode) + '"' in x)]]

  return cpu_node1, cpu_node2, cpu_node3
```

```python
def aggregate_cpu(mode, cpu_node1, cpu_node2, cpu_node3):

  df_cpu_3[str(mode)+'_node3'] = pd.Series(np.random.randint(1, 2, size=len(cpu)))
  for i in cpu_node1.columns:
    df_cpu_3[str(mode)+'_node3'] += cpu[i]

  df_cpu_2[str(mode)+'_node2'] = pd.Series(np.random.randint(1, 2, size=len(cpu)))
  for i in cpu_node2.columns:
    df_cpu_2[str(mode)+'_node2'] += cpu[i]

  df_cpu_1[str(mode)+'_node1'] = pd.Series(np.random.randint(1, 2, size=len(cpu)))
  for i in cpu_node3.columns:
    df_cpu_1[str(mode)+'_node1'] += cpu[i]
```

```python
cpu_modes = ['idle', 'iowait', 'user', 'system', 'irq', 'softirq', 'nice', 'steal']
```

```python
cpu_dfs = []
for n in range(0,7):
  cpu = pd.read_csv('drive/MyDrive/Tesi/Experiment/Data/Prometheus/grafana_data_export_cpu' + str(n) + '.csv', sep=';')
  df_cpu_2 = pd.DataFrame()
  df_cpu_3 = pd.DataFrame()
  df_cpu_1 = pd.DataFrame()
  for i in cpu_modes:
    cpu_node1, cpu_node2, cpu_node3 = get_cpu_metrics_name(i)
    aggregate_cpu(i, cpu_node1, cpu_node2, cpu_node3)

  cpu_dfs.append(pd.concat([cpu['Time'], df_cpu_1, df_cpu_2, df_cpu_3], axis=1))
```

```python
cpu_dfs[4] = cpu_dfs[4][1:]
```

```python
cpu_final = pd.concat([cpu_dfs[0], cpu_dfs[1], cpu_dfs[2], cpu_dfs[3], cpu_dfs[4], cpu_dfs[5], cpu_dfs[6]])
```
